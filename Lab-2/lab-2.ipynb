```python
from matplotlib import pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

fig.patch.set_facecolor('lightgrey')

x = [0, 100]
y = [0, 200]

plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')

ax.plot(x, y)

ax.set_xlim(0,100)
ax.set_ylim(0,200)

ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_yticks([0,50, 100, 150, 200]) 

plt.show()
```

```python
from matplotlib import pyplot as plt

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])

fig.patch.set_facecolor('lightgrey')

x = [0, 100]
y = [0, 200]

ax1.set_title('Title')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax1.plot(x, y)

ax1.set_xlim(0, 100)
ax1.set_ylim(0, 200)
ax1.set_xticks([0, 20, 40, 60, 80, 100])
ax1.set_yticks([0, 50, 100, 150, 200])

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax2.set_title('Title')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax2.plot(x, y)

ax2.set_xlim(0, 100)
ax2.set_ylim(0, 200)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('company_sales_data.csv')

month_numbers = data['month_number']
total_profit = data['total_profit']

fig, ax = plt.subplots()

ax.plot(month_numbers, total_profit, marker='o', color='b', linestyle='-')

ax.set_xlabel('Month Number')
ax.set_ylabel('Total Profit')
ax.set_title('Total Profit Over Months')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('company_sales_data.csv')


month_numbers = data['month_number']
sold_units = data['total_units']


fig, ax = plt.subplots()


ax.plot(month_numbers, sold_units, marker='o', color='red', markerfacecolor='red', 
        linestyle='dotted', linewidth=3, label='Sold Units')


ax.set_xlabel('Month Number')
ax.set_ylabel('Sold Units Number')
ax.set_title('Sold Units Over Months')


ax.legend(loc='lower right')

total=data['total_profit'].sum()

print("Total profit of all months :",total)
plt.show()
```
