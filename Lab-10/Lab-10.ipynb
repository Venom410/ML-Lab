1. Use the IRIS dataset, implement the SVM classifier in python (make use of scikit-learn library), to do the
following.
    
    a. Apply the kernel functions such as linear, polynomial, Radial basis functions and Sigmoid.
    
    b. Plot the scatter plot of the input features.

    c. Plot the decision boundary.

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.inspection import DecisionBoundaryDisplay

# Function to plot the SVM decision boundary for a given kernel
def plot_svm_decision_boundary(X, y, kernel, gamma=0.5, C=1.0):
    # Creating an SVM classifier with the specified kernel, gamma, and C
    svm = SVC(kernel=kernel, gamma=gamma, C=C)
    svm.fit(X, y)  # Fitting the SVM model on the data

    # Creating a DecisionBoundaryDisplay object to visualize the decision boundary
    DecisionBoundaryDisplay.from_estimator(
        svm,
        X,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
    )

    # Plotting the data points with different colors based on their classes
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolors="k")
    plt.title(f'SVM with {kernel} Kernel')
    plt.show()

# Loading the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Take the first two features for visualization
y = iris.target  # Target variable

# List of kernel functions to test
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

# Looping through different kernels and plot the SVM decision boundaries
for kernel in kernels:
    plot_svm_decision_boundary(X, y, kernel)
```
